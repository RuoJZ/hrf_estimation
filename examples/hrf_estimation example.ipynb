## Usage example of hrf_estimation

We will use sample data to estimate the HRF and activation pattern using <code>hrf_estimation</code>

````python
# import packages we will need
%pylab inline
import hrf_estimation as he
````

Now we will load some sample data. This data is described in the paper “Identifying natural images from human brain activity.,” K. N. Kay et al. Nature 2008, and is publicly available from http://crcns.org. 

The following function will download the data to temporary file and perform a local detrending (Savitzky-Golay filter) before returning it.

The returned data consist of an array of 500 voxel timecourses, a list of conditions representing the condition ID and a list of onseets representing the time onset of its relative condition. The list of conditions and onsets have the same length. The repetition time (TR) for this dataset is 1 second.

````python
voxels, conditions, onsets = he.data.get_sample_data(0, full_brain=False)
print('Num of timepoints in a single voxel: %s' % voxels.shape[0])
print('Num of voxels: %s' % voxels.shape[1])
print('Num of conditions: %s' % len(conditions))
print('Num of onsets: %s' % len(onsets))
````

````python
# plot one voxel
import prettyplotlib as ppl
ppl.plot(voxels[:, 0])
plt.xlabel('Timecourse (in seconds)')
plt.show()
````

We will now call the <code>glm</code> function in <code>hrf_estimation</code>. We will use a basis formed by the HRF plus derivatives but a FIR basis is also possible.

````python
# we call he.glm with our data and a TR of 1 second
# we also specify the basis function for the HRF estimation
hrfs, betas = he.glm(conditions, onsets, 1., voxels, basis='3hrf', verbose=1)
````

The output of he.glm produces estimated HRFs and betas. For the case of 3hrf basis these have been normalized so that the resulting HRF correlates positively with the reference HRF and its maximum value is one. 

We will now plot the estimated HRFs,

````python
xx = np.linspace(0, 25) # range of values for time
# construct the final HRF by multiplying by its basis
generated_hrfs = hrfs[0] * he.hrf.spmt(xx)[:, None] + hrfs[1] * he.hrf.dspmt(xx)[:, None] + hrfs[2] * he.hrf.ddspmt(xx)[:, None]
ppl.plot(generated_hrfs)
plt.ylim((-1., 1.))
plt.show()
````

It is possible to use other HRF estimating methods besides R1-GLM. For example, the GLM with separate designs can be used by specifying the keyword argument ```mode='glms'```

````python
hrfs, betas = he.glm(conditions, onsets, 1., voxels, mode='glms', basis='3hrf', verbose=1)
````

Note however that in this case there is an estimation per event and so the shape of ```hrfs``` is ```(n_basis, n_betas, n_voxels)``` and thus must be reduced to ```(n_basis, n_voxels)```. Here we simply average across conditions.

````python
hrfs = hrfs.mean(1)
generated_hrfs = hrfs[0] * he.hrf.spmt(xx)[:, None] + hrfs[1] * he.hrf.dspmt(xx)[:, None] + hrfs[2] * he.hrf.ddspmt(xx)[:, None]
sign = np.sign(np.dot(generated_hrfs.T, he.hrf.spmt(xx)))
norm = np.abs(generated_hrfs).max(0)
generated_hrfs = generated_hrfs * sign / norm
ppl.plot(generated_hrfs)
plt.ylim((-1., 1.))
plt.show()
````
