```python
import pylab as pl
from scipy import ndimage
```

```python
X = np.zeros((5, 3))
X[1, 0] = 1
X[2, 1] = 1
X[0, 2] = 1
```

```python
pl.matshow(X, cmap=pl.cm.Greys, vmax=3.)
pl.ylabel('Time', fontsize=20)
pl.title('Events', fontsize=20)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 5)
pl.text(-0.1, 1.1, '1', fontsize=25)
pl.text(.9, 2.1, '2', fontsize=25)
pl.text(1.9, 0.1, '3', fontsize=25)
# pl.savefig('design1.pdf')
pl.show()
```

```python
e = np.eye(3, 1)
X1 = np.kron(X, e.T)
pl.matshow(X1, cmap=pl.cm.Greys)
pl.show()

X2 = np.zeros_like(X1)
for i in range(X1.shape[0]):
    for j in range(X1.shape[1]):
        X2[i, j] = np.max([X1[i-k, j-k] for k in range(3)])
pl.matshow(X2, cmap=pl.cm.Greys)
pl.show()
```

```python
pl.matshow(X2, cmap=pl.cm.Greys, vmax=3.)
pl.ylabel('Time', fontsize=20)
pl.title('Events', fontsize=20)
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10,5)
pl.text(-0.1, 1.1, '1', fontsize=25)
pl.text(.9, 2.1, '1', fontsize=25)
pl.text(1.9, 3.1, '1', fontsize=25)
pl.text(2.9, 2.1, '2', fontsize=25)
pl.text(3.9, 3.1, '2', fontsize=25)
pl.text(4.9, 4.1, '2', fontsize=25)
pl.text(5.9, 0.1, '3', fontsize=25)
pl.text(6.9, 1.1, '3', fontsize=25)
pl.text(7.9, 2.1, '3', fontsize=25)
# pl.savefig('design2.pdf')
pl.show()
```
